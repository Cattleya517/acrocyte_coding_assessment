## Demo — how `vmin`/`vmax` affect `imshow` display

`imshow` maps array values to greyscale: pixels at `vmax` show as white, pixels at `vmin` (or below) show as black. Same data, different `vmax` → different look. This matters for 12-bit vs 16-bit images later.


```python
im = np.array([
    [0, 1, 2, 1, 0],
    [1, 1, 2, 1, 1],
    [0, 1, 2, 1, 0],
])

fig, ax = plt.subplots(1, 2, figsize=(10, 3))
ax[0].set_title('vmax = 2')
ax[0].imshow(im, cmap='grey', vmin=0, vmax=2)
ax[1].set_title('vmax = 4')
ax[1].imshow(im, cmap='grey', vmin=0, vmax=4)
plt.show()
```
